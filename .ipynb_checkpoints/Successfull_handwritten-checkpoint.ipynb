```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pylab as pl

import pytesseract
pytesseract.pytesseract.tesseract_cmd = 'C:/Program Files/Tesseract-OCR/tesseract.exe'

%matplotlib inline
```

```python
# read
img = cv2.imread(r'C:\Users\ragha\Desktop\maxresdefault.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
def show_image(img):
    pl.gray() 
    pl.matshow(img)
    pl.show() 
    
def show_window(img):
    cv2.imshow('out.jpg', img)
    cv2.waitKey(0)
    cv2.destroyWindow('out.jpg')
```

```python
img = img[0:500, 100:1280]
show_image(img)
```

```python
def detect(c):
    # initialize the shape name and approximate the contour
    shape = "unidentified"
    peri = cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c, 0.04 * peri, True)
    if len(approx) == 4:
        return True
    return False
```

```python
def increase_contrast(img):
    # increase contrast
    pxmin = np.min(img)
    pxmax = np.max(img)
    imgContrast = (img - pxmin) / (pxmax - pxmin) * 255

    # increase line width
    kernel = np.ones((3, 3), np.uint8)
    return cv2.erode(imgContrast, kernel, iterations = 1)
```

```python
img = increase_contrast(img)
```

```python
show_image(img)
```

```python
blurred = cv2.GaussianBlur(img, (5, 5), 0)
binary_thresh = cv2.threshold(blurred, 110, 255, cv2.THRESH_BINARY)[1]
```

```python
show_window(binary_thresh)
```

```python
text = pytesseract.image_to_string(binary_thresh)
```

```python
show_window(binary_thresh)
```
